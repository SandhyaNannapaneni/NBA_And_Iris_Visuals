A pair plot for the Iris dataset was visualized using Seaborn to make comparisons of features and group them by types of species. After the dataset has been successfully imported to a Pandas data frame, the pair plot produces numerous scatterplots each for a combination of the features, then the dots are appropriately colored based on the species to which the data points belong. On the positive side of this cover, the diagonal contains Kernel Density Estimations (KDE) to determine the distribution of a different feature. Such a chart serves as a tool for detecting phenomena well, say more about the unpolluted discreteness of the Setosa species of iris and the distribution of hearts Among the Irises Versicolor and Virginica. Thus, the chart is excellent for giving the geometrical representation of the information.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# loading the Iris dataset
file_path = 'iris.csv'  
sndata = pd.read_csv(file_path)

# Creating a customized pair plot for the Iris dataset
sns.pairplot(
    sndata, 
    hue='species', 
    diag_kind='kde', 
    palette={'setosa': '#F4D03F', 'versicolor': '#884EA0', 'virginica': '#66FFCC'},  # New custom colors
    vars=['sepal_width', 'petal_width', 'sepal_length']  # New selected variables for the plot
)
plt.suptitle("Customized Pair Plot of New Features in the Iris Dataset", y=1.02, fontsize=16)
plt.show()
```

QUESTION : 2

The average petal length of each iris species in the Iris data set is returned and visualized using an easy-to-understand bar plot. The group function arranges the data in species and calculates the petal length for every species. In this case, there is an accurate and well-detailed bar chart with custom colors, distinct axes, and titles. The axis range is defined for clarity purposes while the plots are clear and have the crucial elements: displays the bars on the same grid, thus making it easy to visually divide the different species of iris. The grid on the plot itself puts the information in sectors for a more decipherable view and as well clearly shows the disparity in the length of the petals on the three kinds, which are Setosa, Versicolor, and Virginia. This is a typical example of bar markers plotting the values of petal length differences of the species.

```python
# Calculate the average petal length for each species
avg_petal_length = sndata.groupby('species')['petal_length'].mean()

# Create a customized bar plot
plt.figure(figsize=(8, 6))
colors = ['#F4D03F', '#884EA0', '#66FFCC']  # Custom color palette
plt.bar(avg_petal_length.index, avg_petal_length, color=colors)
plt.title('Average Petal Length by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Average Petal Length (cm)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The dimensionality of the Iris dataset was two thanks to the Principal component Analysis. Therefore,  a certain amount of information has been lost during this procedure although the aim was focused on only two dimensions; only two components. Features were extracted and data was transformed using PCA into a  2D space in the–X – Y axis, then displayed as a scatter plot. displayed and dots are color-coded in terms of distinct species in the data. The scatter plot showcases the clustering of ‘Setosa’ clearly and ’Versicolor’ and ‘Virginica’ having an overlap. This simplifies the dataset externally but maintains all the important relationships within the data.

```python
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

# Perform PCA on the Iris dataset
iris_features = sndata.drop(columns=['species'])
species_labels = sndata['species']

pca = PCA(n_components=2)
pca_result = pca.fit_transform(iris_features)

# Create a scatter plot for PCA results
pca_df = pd.DataFrame(pca_result, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['species'] = species_labels

# Define a color palette for the species
colors = {'setosa': '#F4D03F', 'versicolor': '#884EA0', 'virginica': '#66FFCC'}

plt.figure(figsize=(10, 8))
for species in pca_df['species'].unique():
    subset = pca_df[pca_df['species'] == species]
    plt.scatter(
        subset['Principal Component 1'], 
        subset['Principal Component 2'], 
        label=species, 
        color=colors[species],  # Use the defined colors
        alpha=0.7
    )
plt.title('PCA: First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Species')
plt.grid()
plt.tight_layout()
plt.show()
```

This applies t-SNE (t-distributed Stochastic Neighbor Embedding) on the Iris dataset and aims to optimize the parameters in such a manner that the overall computational complexity is reduced considerably. t-SNE allows for the reduction of the data but without the distortion of the local configuration of the data. The transformed data in this case is shown in a plot form, a scatter plot with points shown as color-coded as per the respective species. The plotted data thus effectively reveals isolated groups, easily recognizable by the separation of Setosa but the inter-stage intersection of seasoned meat and Versicolor and Virginica. This representation is more effective for dissecting high-dimensional data and extracting all nonlinear and interactive patterns or cluster structures in it.

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pandas as pd

# Retry t-SNE with optimized parameters for reduced complexity
tsne_optimized = TSNE(n_components=2, random_state=42, perplexity=10, n_iter=250, init='random')

try:
    tsne_result_optimized = tsne_optimized.fit_transform(iris_features)

    # Create a DataFrame for optimized t-SNE results
    tsne_df_optimized = pd.DataFrame(tsne_result_optimized, columns=['t-SNE Component 1', 't-SNE Component 2'])
    tsne_df_optimized['species'] = species_labels

    # Define a color palette for the species
    colors = {'setosa': '#F4D03F', 'versicolor': '#884EA0', 'virginica': '#66FFCC'}

    # Plot optimized t-SNE results
    plt.figure(figsize=(10, 8))
    for species in tsne_df_optimized['species'].unique():
        subset = tsne_df_optimized[tsne_df_optimized['species'] == species]
        plt.scatter(
            subset['t-SNE Component 1'], 
            subset['t-SNE Component 2'], 
            label=species, 
            color=colors[species],  # Use the defined colors
            alpha=0.7
        )
    plt.title('t-SNE: Optimized Visualization of Iris Dataset')
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.legend(title='Species')
    plt.grid()
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Optimized t-SNE failed: {e}")
```

compares the PCA and t-SNE visualizations for the Iris dataset, representing their differences in dimensionality reduction techniques. PCA reduces the data into two components, which represent maximal variance and thus retention of global structure, while t-SNE focuses on local preservation of relationships, hence applicable to cluster separability. Both visualizations were created with the same color codes for the species: Setosa in blue, Versicolor in red, and Virginica* in green. The scatter plots shown carry appropriate labels on the axis, legends, and grid lines. PCA presents a fast interpretation, while t-SNE renders cluster shapes better, especially for non-linear relationships. This comparison illustrates the respective advantages of PCA and t-SNE in visualizing high-dimensional data. 

```python
# Comparing the PCA Visualization and t-SNE Visualization
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Loading the Iris dataset
file_path = 'iris.csv'
sndata = pd.read_csv(file_path)

# Preparing the data
iris_features = sndata.drop(columns=['species'])
species_labels = sndata['species']

colors = {'setosa': '#F4D03F', 'versicolor': '#884EA0', 'virginica': '#66FFCC'}

# PCA Visualization

pca = PCA(n_components=2)
pca_result = pca.fit_transform(iris_features)

# Creating a DataFrame for PCA results
pca_df = pd.DataFrame(pca_result, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['species'] = species_labels

# Plot PCA results
plt.figure(figsize=(10, 8))
for species in pca_df['species'].unique():
    subset = pca_df[pca_df['species'] == species]
    plt.scatter(
        subset['Principal Component 1'], 
        subset['Principal Component 2'], 
        label=species, 
        color=colors[species], 
        alpha=0.7
    )
plt.title('PCA: First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Species')
plt.grid()
plt.tight_layout()
plt.show()

# t-SNE Visualization

# Perform t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=10, n_iter=250, init='random')
tsne_result = tsne.fit_transform(iris_features)

# Create a DataFrame for t-SNE results
tsne_df = pd.DataFrame(tsne_result, columns=['t-SNE Component 1', 't-SNE Component 2'])
tsne_df['species'] = species_labels

# Plot t-SNE results
plt.figure(figsize=(10, 8))
for species in tsne_df['species'].unique():
    subset = tsne_df[tsne_df['species'] == species]
    plt.scatter(
        subset['t-SNE Component 1'], 
        subset['t-SNE Component 2'], 
        label=species, 
        color=colors[species], 
        alpha=0.7
    )
plt.title('t-SNE: Optimized Visualization of Iris Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend(title='Species')
plt.grid()
plt.tight_layout()
plt.show()
```
